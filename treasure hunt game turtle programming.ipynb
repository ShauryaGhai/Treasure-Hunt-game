```python
#import all libraries
import turtle
import random
import math
#create turtle t on screen
t=turtle.Turtle()
#Hide treasure
treasure_x=random.randint(-100,100)
treasure_y=random.randint(-100,100)
print(treasure_x)
print(treasure_y)
#Enter current_x and current_y coordinate value and store in previous_x and previous_y. 
current_x=int(input('Enter current guess X coordinate lies between -100 and 100'))
current_y=int(input('Enter current guess Y coordinate lies between -100 and 100'))
previous_x=current_x
previous_y=current_y
#create variable gameover
gameover=False
#Write your while loop code here
while gameover==False:
#Enter again current position of user and store in current_x and current_y 
  current_x=int(input('Enter current guess X coordinate lies between -100 and 100'))
  current_y=int(input('Enter current guess Y coordinate lies between -100 and 100'))
  #Move the turtle to user current position
  t.goto(current_x,current_y)
  #Calculate previousdistance
  previousdistance=math.sqrt(((previous_x-treasure_x)**2)+((previous_y-treasure_y)**2))
  #calculate currentdistance
  currentdistance=math.sqrt(((current_x-treasure_x)**2)+((current_y-treasure_y)**2))
  #Now check condition by comparing currentdistance and previousdistance
  if current_x == treasure_x and current_y ==treasure_y:
    t.color('red')
    t.write('you won the game')
    gameover=True
  elif currentdistance <previousdistance:
    t.color('orange')
    t.write('you are getting warmer')
  elif currentdistance > previousdistance:
    t.color('dark blue')
    t.write('you are getting colder')
  

```
